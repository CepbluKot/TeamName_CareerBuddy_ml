```python
import pandas as pd
import warnings
import pickle
from sklearn.preprocessing import StandardScaler


warnings.filterwarnings("ignore")

df = pd.read_csv("dataset.csv")


random_input = df.iloc[[34]]
random_input

# random_input.to_dict()
df = pd.concat([df, pd.DataFrame(random_input)])
# print(random_input)
new_index = list(df.index)
new_index[-1] = len(new_index)
df.index=new_index
df
```

```python

cat_cols = [
    "Attrition",
    "BusinessTravel",
    "Department",
    "Education",
    "EducationField",
    "EmployeeCount",
    "EnvironmentSatisfaction",
    "Gender",
    "JobInvolvement",
    "JobLevel",
    "JobRole",
    "JobSatisfaction",
    "MaritalStatus",
    "Over18",
    "OverTime",
    "PerformanceRating",
    "RelationshipSatisfaction",
    "StandardHours",
    "StockOptionLevel",
    "WorkLifeBalance",
]

for i in df[cat_cols]:
    if i == "Attrition":
        continue
    df[i] = pd.Categorical(df[i])

df.drop("EmployeeNumber", axis=1, inplace=True)
```

```python

vars_rm = [
    "Education",
    "EmployeeCount",
    "Gender",
    "Over18",
    "PerformanceRating",
    "RelationshipSatisfaction",
    "StandardHours",
]

df = df.loc[:, df.columns != "Attrition"]
df.drop(df[vars_rm], axis=1, inplace=True)
```

```python

dummy_cols = [
    "OverTime",
    "JobRole",
    "JobLevel",
    "StockOptionLevel",
    "MaritalStatus",
    "JobInvolvement",
    "BusinessTravel",
    "EnvironmentSatisfaction",
    "JobSatisfaction",
    "WorkLifeBalance",
    "Department",
    "EducationField",
]


df = pd.get_dummies(df, columns=dummy_cols, drop_first=True)
df.columns = df.columns.str.replace(" ", "")
```

```python
scaler = StandardScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python

rf_cv = None
with open('model.pkl', 'rb') as file:
    rf_cv = pickle.load(file)

predicted_row = df.iloc[-1].values.reshape(1, -1)  # Изменение формы для предсказания
prediction = rf_cv.predict(predicted_row)


# print(predicted_row)
print(prediction)
```
